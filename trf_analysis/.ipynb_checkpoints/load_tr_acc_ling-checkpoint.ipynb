```python
import mne 
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import scipy
```

```python
#Load unshuffled acc
subjects = ['sub05','sub06', 'sub07', 'sub08','sub09','sub10','sub11','sub12','sub13','sub14','sub15','sub16',
'sub18','sub19','sub20','sub21','sub22','sub23', 'sub25', 'sub26', 'sub27', 'sub28', 'sub29', 'sub30', 
'sub31', 'sub32', 'sub33', 'sub34', 'sub35', 'sub36', 'sub37', 'sub38', 'sub39', 'sub40', 'sub41', 'sub42', 'sub43', 
'sub44', 'sub45', 'sub46', 'sub47', 'sub48', 'sub49', 'sub50', 'sub51', 'sub52', 'sub53', 'sub55', 'sub56', 
'sub57', 'sub58', 'sub59', 'sub60', 'sub61', 'sub62',  
'sub64', 'sub65', 'sub66', 'sub67', 'sub68', 'sub69', 'sub70', 'sub72']


acctr = []
for subject in subjects:
    print(subject)
    acc_unshuff = pd.read_csv("/data/pt_03123/trf_results_extra/{s}/pad_multi/fits_linguistic_ws_{s}_within_nb_stan.csv".format(s=subject))
    
    
    acc_shuff = []
   # feats = ['ac_target', 'ac_dis', 'onsets_target', 'onsets_dis']
#    feats = [ 'ling_word_target', 'ling_word_dis', 'ling_phone_target', 'ling_phone_dis']
    feats = [ 'ling_word_target']

    #Load shuffled acc
    numit = 50
    for feat in feats:
        for it in range(1,numit):
            shuff = pd.read_csv("/data/pt_03123/trf_results_extra/{s}/shuffle/fits_linguistic_pad_ws_{s}_{i}_{f}_within_nb_stan.csv".format(s = subject, i = it, f = feat))
            shuff['feature'] = shuff['feature_shuffle']
            shuff['subject'] = subject
            acc_shuff.append(shuff)
    
    acc_shuff = pd.concat(acc_shuff)
    
    acc_shuff = acc_shuff.groupby(['trialnum', 'channel', 'window', 'feature', 'subject'], as_index = False).mean('fit')              
    
    #Merge 
    accmere = pd.merge(acc_unshuff, acc_shuff, on = ['subject', 'trialnum', 'channel', 'window', 'feature'])
    accmere['fitsub'] = accmere['fit_x'] - accmere['fit_y'] 

    acctr.append(accmere)
```

```python
acctr = pd.concat(acctr)
```

```python
from mne.stats import spatio_temporal_cluster_1samp_test

sigres = []
subjectacc_sig = []
for f,fdf in acctr.groupby('feature'):
    featac = []
    print(f)
    datalist = []
    for s,sdf in fdf.groupby('subject'):
       # print(s)
        dfm = sdf.groupby(['channel', 'window'], as_index = False, sort = False)['fitsub'].mean()

        channel_order = dfm['channel'].unique()
        dfm['channel'] = pd.Categorical(dfm['channel'], categories=channel_order, ordered=True)

        pivoted = dfm.pivot(index='window', columns='channel', values='fitsub')

        d = pivoted.to_numpy()
        datalist.append(d)

    datalist = np.array(datalist)
    print(datalist.shape)

    #Do significance testing 

    #Create adjacency 
    subject="sub05"
    eeg_path = "./eeg_mne/saved_{sub}_new-epo.fif".format(sub = subject)
    
    ep = mne.read_epochs(eeg_path)
    ep = ep.drop_channels(['A1', 'A2'])

    channels = dfm['channel'].unique()
    
    ep_filt = ep.copy().pick(list(channels))
    ad,ch= mne.channels.find_ch_adjacency(ep_filt.info, ch_type = "eeg")
    ad = mne.stats.combine_adjacency(datalist.shape[1], ad)

    res = spatio_temporal_cluster_1samp_test(
    datalist,
   # threshold = dict(start = 0, step = 0.2),
    verbose=True, adjacency = ad, seed = 1, n_jobs = 5,
        tail = 1
    )


    cor_mean =np.mean(np.array(datalist), axis = 0)
    incor_mean = np.mean(np.array(datalist), axis = 0)
    diff_trf = cor_mean-incor_mean
    
    t_obs, clusters, pvals, h0 = res
    cluster_alpha = 0.05
    print(len(res[1]))
    #print(pvals)
    ncs = 0
    if len(res[1]) > 0:
            # no idea what is happening here
        nan_trf = np.ones_like(cor_mean) * np.nan
        nan_trf = np.ones_like(cor_mean) * np.nan
        nan_trf_list = np.ones_like(cor_mean) * np.nan
        nan_diff = nan_trf.copy()
        ncs = 0
        for k_cluster, c in enumerate(clusters):
            
          
            if pvals[k_cluster] < cluster_alpha:
                ncs = ncs+1
                lags = c
                ct = c[c == True]
               # print(len(set(ct)))
                
                if len(set(ct)) > 0:
                    if abs(cor_mean[lags].mean()) > abs(incor_mean[lags].mean()):
                        nan_trf[lags] = cor_mean[lags]
                    else:
                        nan_trf_list[lags] = incor_mean[lags]
                    nan_diff[lags] = diff_trf[lags]
                    
                signi_times_t = (~np.isnan(nan_diff)).astype(float)
                signi_times_t[signi_times_t == 0.] = np.nan
                signi_chans_bool_t = (~np.all(np.isnan(nan_diff), axis=0))
               # print(signi_times_t)
        
            
    else:
        signi_times_t = diff_trf
        signi_chans_bool_t = np.nan
    
    if ncs == 0:
        signi_times_t = diff_trf
        signi_chans_bool_t = np.nan

    sigres.append(signi_times_t)
    
    #Filter the original dfs for significant sensors & channels 

    for sar in datalist:
        
        bool_mask = signi_times_t == 1
        print(bool_mask)
        #Filter only for time 
        time_points_with_true = np.any(bool_mask, axis=1)

        # Set all channels to True for those time points
        bool_mask[time_points_with_true] = True
        
        
        
        # Apply the mask to filter the data array
        filtered = np.where(bool_mask, sar, np.nan)
        subac = np.nanmean(filtered)

        featac.append(subac)

    subjectacc_sig.append(featac)
        

        
```

```python
def make_ageplot(feat, acctr, plot_scatter = False, window="late", windowstart = 0.4, windowend = None):

    dat = acctr.copy()
    
    datsub = dat[dat['feat'] == feat].groupby(['window', 'subject', 'ageg'], as_index = False).mean('fitsub')
    
    stan = datsub.groupby(['ageg', 'window'], as_index = False)['fitsub'].sem()
    
    stan = stan.rename(columns={"fitsub": "std"})
    
    dat = acctr.copy()
    
    dat = dat[dat['feat'] == feat].groupby(['window', 'ageg'], as_index = False).mean('fitsub')
    
    dat_std = pd.merge(dat, stan, on = ['ageg', 'window'])

    #Get correlation for late window
    if window == "late":
        datlate = datsub[datsub['window'] > 21]
        datlate=datlate.groupby(['subject', 'Age', 'ageg'], as_index = False)['fitsub'].mean()
        print(datlate[['Age', 'fitsub']].corr().iloc[0,1])
    else:
        datlate = datsub[(datsub['window'] > (windowstart+0.1)*(128/3)) & (datsub['window'] < ((windowend+0.1)*(128/3)))]
        datlate=datlate.groupby(['subject', 'Age', 'ageg'], as_index = False)['fitsub'].mean()
        c = scipy.stats.pearsonr(datlate['Age'], datlate['fitsub'])
        print(c.statistic)

        cor = np.round(c.statistic, 2)
        p = np.round(c.pvalue,3)


    coldict={'young':'#98accf', 'middle':'#5579aa','old':'#3f4756'}


    x = np.arange(-0.1, 0.78, 1/128*3)
    # Define the sorter
    sorter = ['young', 'middle', 'old']

    
    dat_std.ageg = dat_std.ageg.astype("category")
    dat_std.ageg = dat_std.ageg.cat.set_categories(sorter)

    
    if plot_scatter == True:
        fig,ax=plt.subplots(figsize=[10,3], ncols = 2, width_ratios=[0.7, 0.3])
        
        
        for key, group in dat_std.groupby('ageg'):
            group['x'] = x
            ax[0].plot(group['x'], group['fitsub'], label=key, color = coldict[key], linewidth = 3)
            ax[0].fill_between(group['x'],  group['fitsub'] + group['std'], group['fitsub'] - group['std'], color = coldict[key], 
                               alpha = .15)
        
        ax[0].legend()
        ax[0].set_xlabel("Time (s)",size=20)
        ax[0].set_ylabel("Encoding\naccuracy",size=20)
        
        ax[0].tick_params(axis='both', which='major', labelsize=15)
        
        ax[0].hlines(y=-0.0022, color='grey', linestyle='-', linewidth = 5, alpha = 0.5, xmin = windowstart, xmax = windowend)
        
        
        for key, group in datlate.groupby('ageg'):
            ax[1].scatter(group['Age'], group['fitsub'], label=key, color = coldict[key])
        
        ax[1].set_xlabel("Age",size=20)
        ax[1].tick_params(axis='both', which='major', labelsize=15)
        ax[1].annotate("r = {c}, p = {p}".format(c=cor, p = p), xy = (20, 0.03), size = 15)
        fig.tight_layout()
        
        plt.savefig("../figures/tr_encoding_{f}.svg".format(f=feat), bbox_inches = "tight")

    else:
        fig,ax=plt.subplots(figsize=[7,3], ncols = 1)
        
        
        for key, group in dat_std.groupby('ageg'):
            ax.plot(group['x'], group['fitsub'], label=key, color = coldict[key], linewidth = 3)
            ax.fill_between(group['x'],  group['fitsub'] + group['std'], group['fitsub'] - group['std'], color = coldict[key], 
                               alpha = .15)
        
        ax.legend()
        ax.set_xlabel("Time (s)",size=20)
        ax.set_ylabel("Encoding\naccuracy",size=20)
        
        ax.tick_params(axis='both', which='major', labelsize=15)
        
       # ax.axhline(y=-0.0022, color='grey', linestyle='-', linewidth = 5, alpha = .5, xmin = 0.57, xmax = 0.96)
        
       
        fig.tight_layout()
        
        plt.savefig("../figures/tr_encoding_{f}.svg".format(f=feat), bbox_inches = "tight")
```
